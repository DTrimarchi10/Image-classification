Using TensorFlow Datasets

## Import Libraries

```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
#import numpy as np
#from PIL import Image
```

```python
tf.enable_eager_execution()
tf.executing_eagerly()
```

## Load cifar10 database

```python
datasets, info = tfds.load("cifar10", with_info=True)
train_dataset, test_dataset = datasets["train"], datasets["test"]
assert isinstance(train_dataset, tf.data.Dataset)
print(info)
```

View features

```python
print(info.features)
print(info.features["label"].num_classes)
print(info.features["label"].names)
```

See example images from dataset

```python
fig = tfds.show_examples(info, train_dataset)
```

```python
for example in train_dataset.take(1):  # Only take a single example
    plt.imshow(example['image'].numpy())
```
